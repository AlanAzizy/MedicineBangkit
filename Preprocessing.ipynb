```python
import pandas as pd

# Replace 'file.xlsx' with the path to your Excel file
file_path = '/content/Drug Database.xlsx'

# Read the Excel file, specifying sheet_name=0 to load the first sheet
df = pd.read_excel(file_path, sheet_name=0)
df_1 = pd.read_excel(file_path, sheet_name=1)
df_2 = pd.read_excel(file_path, sheet_name=2)
df = pd.concat([df, df_1, df_2])
```

```python
df.head()
```

```python
df['index'] = range(1, len(df) + 1)
```

```python
df = df.drop(columns=['id','Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10','ID','This data is collected from malacards.org'])
```

```python
df = df.sort_values(by='Drug')
df.dropna()
df
```

```python
df = df.replace({r'\s+$': '', r'^\s+': ''}, regex=True).replace(r'\n',  ' ', regex=True)
df = df.sort_values(by='Drug')
df
```

```python
unik_value = df.groupby('Drug').count().reset_index()
unik_value
```
